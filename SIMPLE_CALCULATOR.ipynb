```python
import math

def add(x, y): return x + y
def subtract(x, y): return x - y
def multiply(x, y): return x * y
def divide(x, y):
    if y == 0:
        print("Cannot divide by zero.")
        return 0
    return x / y
def modulus(x, y): return x % y
def exponent(x, y): return x ** y
def floor_division(x, y): return x // y
def square_root(x): return math.sqrt(x)

def multiple_calculations():
    print("Multiple Calculations Mode")
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Modulus")
    print("6. Exponent")
    print("7. Floor Division")
    print("-1. Exit")

    first_time = True
    ans = 0

    while True:
        choice = input("\nEnter choice: ")

        if choice == '-1':
            print(f"Final answer: {ans}")
            break

        if first_time:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
        else:
            num1 = ans
            num2 = float(input("Enter next number: "))

        if choice == '1':
            ans = add(num1, num2)
        elif choice == '2':
            ans = subtract(num1, num2)
        elif choice == '3':
            ans = multiply(num1, num2)
        elif choice == '4':
            if num2 != 0:
                ans = divide(num1, num2)
            else:
                print("Can't divide by zero.")
                continue
        elif choice == '5':
            ans = modulus(num1, num2)
        elif choice == '6':
            ans = exponent(num1, num2)
        elif choice == '7':
            ans = floor_division(num1, num2)
        else:
            print("Invalid input. Try again.")
            continue

        print(f"Result: {ans}")
        first_time = False

def calculator():
    print("Simple Calculator")
    print("1. Add\n2. Subtract\n3. Multiply\n4. Divide\n5. Modulus\n6. Exponent\n7. Floor Division\n8. Square Root\n9. Multiple Calculations\n-1. Exit")
    while True:
        choice = input("\nEnter choice: ")
        if choice == '-1':
            print("Exiting Calculator. Goodbye!")
            break
        elif choice in ['1','2','3','4','5','6','7']:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
            if choice == '1': print(f"{num1} + {num2} = {add(num1, num2)}")
            elif choice == '2': print(f"{num1} - {num2} = {subtract(num1, num2)}")
            elif choice == '3': print(f"{num1} * {num2} = {multiply(num1, num2)}")
            elif choice == '4': print(f"{num1} / {num2} = {divide(num1, num2)}")
            elif choice == '5': print(f"{num1} % {num2} = {modulus(num1, num2)}")
            elif choice == '6': print(f"{num1} ** {num2} = {exponent(num1, num2)}")
            elif choice == '7': print(f"{num1} // {num2} = {floor_division(num1, num2)}")
        elif choice == '8':
            num = float(input("Enter number: "))
            print(f"âˆš{num} = {square_root(num)}")
        elif choice == '9':
            multiple_calculations()
        else:
            print("Invalid input. Please select a valid operation.")

calculator()
```
